```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
classificador = LinearRegression()
scaler_standard = StandardScaler()
```

```python
previsores = scaler_standard.fit_transform(load_boston()['data'])
classe = load_boston()['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(previsores,classe,
                                                    test_size=0.15, 
                                                    random_state=12)
```

```python
def classifica(modelo, scaler, previsor, classe):
    classificador = modelo()
    scaler = scaler()
    X_train, X_test, y_train, y_test = train_test_split(previsor,classe,
                                                    test_size=0.20, 
                                                    random_state=12)
    classificador.fit(X_train, y_train)
    print('Score on Test: ', classificador.score(X_test,y_test))
    print('Score on Train: ', classificador.score(X_train,y_train))
```

```python
classificador.fit(X_train, y_train)
print(classificador.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100)

rf.fit(previsores, classe)
print(classificador.score(previsores,classe))
```

```python
rf.feature_importances_
```

```python
def featureImportances(rfInstance, X_train_columns):
   feature_importance = pd.DataFrame(rfInstance.feature_importances_,\
      index = X_train_columns, columns=['importance'])\
         .sort_values('importance', ascending=False)
   return feature_importance  
```

```python
featureImportances(rf, load_boston()['feature_names'])
```

```python
result_train = []
result_test = []
alpha = np.arange(0,5,0.2)

for a in alpha:
    rr = Ridge(alpha=a)
    rr.fit(X_train, y_train)
    result_train.append(rr.score(X_train,y_train))
    result_test.append(rr.score(X_test,y_test))
    
plt.plot(alpha,result_train, 'g-')
plt.plot(alpha,result_test, 'r-')
plt.show()
```
